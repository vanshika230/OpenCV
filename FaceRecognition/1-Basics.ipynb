```python
import cv2
```

```python
# Read an Image
img = cv2.imread('me.jpg')
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img) #default is in bgr format
plt.show()
```

```python
newImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newImg)  #rgb format - original photo
plt.show()
```
